# Plotting in Pandas

```python
n = pd.Series(np.random.normal(size= 200))
n.plot()
print(n)
```

```python
n = pd.Series(np.random.normal(size= 30))
print(n)
n.plot(grid = True)
```

```python
n.cumsum().plot(grid= True)
```

```python
cv = pd.DataFrame({'normal':np.random.normal(size = 10),
                         'Gamme':np.random.gamma(1,size=10),
                         'Poisson':np.random.poisson(size = 10)})
cv.cumsum(0).plot()
print(cv)
```

```python
cv.cumsum(0).plot(subplots =True)
```

```python
cv.cumsum(0).plot(secondary_y='normal')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(4, 12))
for i,var in enumerate(['normal','Gamme','Poisson']):
    cv[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

# Simple Line Plots

# Line Colors and Styles

```python
plt.plot(a, np.sin(a - 0), color='blue')        
plt.plot(a, np.sin(a - 1), color='y')           
plt.plot(a, np.sin(a - 3), color='0.55')    
```

```python
plt.plot(a, np.sin(a - 0), linestyle='-')        
plt.plot(a, np.sin(a - 1), linestyle=':')           
plt.plot(a, np.sin(a - 2), linestyle='-.')  
```

# Adjusting the plot: axes limits

```python
plt.plot(a, np.sin(a - 0))

plt.xlim(-1, 20)
plt.ylim(-2, 10);  #limiting the y axis and the  x axis
```

```python
rng = np.random.RandomState(7)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.6,cmap='Greens')
plt.colorbar()
```

# Visualizing errors

### Errorbars

```python
x = np.linspace(0, 25, 100)
dy = 0.1
y = np.sin(x) + dy * np.random.randn(100)

plt.errorbar(x, y, yerr=dy, fmt='-d');
```

```python
x = np.linspace(0, 25, 100)
dy = 0.1
y = np.tan(x) + dy * np.random.randn(100)

plt.errorbar(x, y, xerr=dy, fmt='-d');
```

# Histogram

```python
plt.style.use('seaborn-white')

data = np.random.randn(10000)
print(data)
```

```python
plt.hist(data)
```

```python
plt.hist(data, bins=10, normed=True, alpha=0.6,
         histtype='stepfilled')
```

```python
x = np.random.normal(0, 0.8, 1000)
y = np.random.normal(-2, 1, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.5, normed=True, bins=50)

plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
```

```python
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

# Two_dimensional histograms

```python
mean = [8,10]
cov = [[5,4], [2, 0.5]]
x, y = np.random.multivariate_normal(mean, cov, 10000)
```

# plt.hist2d:two-dimensional histogram

```python
plt.hist2d(x, y, bins=50, cmap='Greens')
cb = plt.colorbar()
```

# PLT.HEXBIN

```python
plt.hexbin(x, y, gridsize=10, cmap='Blues')

cb = plt.colorbar()
```
